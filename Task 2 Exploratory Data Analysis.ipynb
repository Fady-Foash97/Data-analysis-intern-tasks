## Now we are drawing histogram

```python
price_data = df[["open", "close", "high", "low"]]
```

```python
plt.figure(figsize=(10,7))
plt.hist(price_data, bins=10)
plt.title("Price stocks distribution")
plt.show()
```

## Distribution of price data of stocks are closer to each other, there is no difference between prices:

```python
volume_data = df[["volume"]]
```

```python
plt.hist(volume_data, bins=10)
plt.title("Volume distribution")
plt.show()
```

## Distribution of volume is higher than price

## Boxplot:

```python
figure = plt.figure(figsize =(14, 14))
sns.boxplot(price_data)
plt.show()
```

## High price distribution is slightly higher than open, close and low prices
