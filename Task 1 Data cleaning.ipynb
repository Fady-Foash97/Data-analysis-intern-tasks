## Task 1:
-In this task, we are going to practice data cleaning.

```python
import pandas as pd
```

```python
df = pd.read_csv("3) Sentiment dataset.csv")
```

```python
df
```

## Types of columns

```python
df.info()
```

## Change "Retweets" and "likes" columns types from float to integer

```python
df = df.astype({"Retweets": int, "Likes": int})
```

```python
df.info()
```

## Change "Timestamp" column type from object to datetime

```python
df["Timestamp"] = pd.to_datetime(df["Timestamp"])
```

```python
df.info()
```

## Drop "Unnamed: 0.1", "Unnamed: 0", "Year", "Month", "Day", "Hour" columns

```python
df_droped_columns = df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0', 'Year', 'Month', 'Day', 'Hour'])
```

```python
df_droped_columns
```

```python
df_droped_columns.info()
```

## Sort columns by "Timestamp" column in ascending order

```python
df_sort_time = df_droped_columns.sort_values(by = "Timestamp", ascending = True)
```

```python
df_sort_time
```

## Reset index to arrange them in ascending order

```python
df_index = df_sort_time.reset_index()
```

```python
df_index
```

## We have a problem! An "index" column is created out of nowhere, now we are going to drop this column

```python
df_sort_index = df_index.drop(columns=['index'])
```

```python
df_sort_index
```

## Now we need to process the data to find out if there is data that need to be cleaned or not

```python
df_sort_index.isnull().sum()
```

## There are no nulls

```python
df_sort_index.duplicated().sum()
```

## There are 20 duplicated data

```python
df_drop_duplicate = df_sort_index.drop_duplicates()
```

```python
df_drop_duplicate
```

## There are 712 rows now, we need to reset index

```python
df_sort_index1 = df_drop_duplicate.reset_index()
```

```python
df_sort_index1
```

```python
df_sort_index2 = df_sort_index1.drop(columns=['index'])
```

```python
df_sort_index2
```

```python
df_sort_index2['Country'].value_counts().reset_index()
```

## Japan and USA is repeated 2 times

## We are going to combine the USA and Japan repeated rows with each other by removing any line spaces between variations of countries

```python
df_sort_index2['Country'] = df_sort_index2['Country'].str.strip() 
```

```python
df_sort_index2['Country'].value_counts().reset_index()
```
